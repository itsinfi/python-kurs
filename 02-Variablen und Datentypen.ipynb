### Dateien

In Python mÃ¼ssen Sie dafÃ¼r zunÃ¤chst einmal die vorgesehene Datei mit der Methode open Ã¶ffnen, 
wobei neben der Angabe des Dateinamens auch die Wahl des Bearbeitungsmodus erforder-
lich ist. In Bezug auf den Letzteren haben Sie die Wahl zwischen den Optionen "x" (Erst-
ellungsmodus), "a" (AnfÃ¼gemodus), "w" (Schreibmodus) und "r" (Lesemodus):

- Die Option **"r" (Read)** Ã¶ffnet die angegebene Datei im Lesemodus. Da es sich hier
um die Standardeinstellung handelt, muss sie nicht explizit angegeben werden.
- Die Option **"x" (Create)** veranlasst die Erstellung einer neuen Datei mit dem angege-
benen Namen. Falls bereits eine Datei mit dieser Bezeichnung existiert, wird eine
Fehlermeldung ausgegeben.
- Die Option **"a" (Append)** bewirkt die Erstellung einer neuen Datei, falls der angege-
bene Dateiname nicht existiert. Wenn bereits eine Datei mit der genannten Bezeich-
nung vorhanden ist, wird diese Datei geÃ¶ffnet â ohne Ausgabe einer Fehlermeldung.
AnschlieÃend werden alle neuen EintrÃ¤ge am Ende der Datei angefÃ¼gt, sodass die
bereits bestehenden Dateiinhalte erhalten bleiben.
- Die Option **"w" (Write)** bewirkt ebenfalls die Erstellung einer neuen Datei, falls der
angegebene Dateiname nicht existiert. Wenn bereits eine Datei mit der genannten
Bezeichnung vorhanden ist, wird diese Datei geÃ¶ffnet (wiederum ohne Ausgabe
einer Fehlermeldung). Allerdings werden in letzterem Fall â anders als im AnfÃ¼gemo-
dus â alle vorherigen Inhalte der bereits bestehenden Datei gelÃ¶scht. Das bedeutet
auch, dass alle neuen EintrÃ¤ge am Dateianfang eingefÃ¼gt werden.

In jedem der vier FÃ¤lle gibt die open-Methode eine Referenz auf ein Dateiobjekt aus,
die Sie fÃ¼r alle folgenden Zugriffsoperationen benÃ¶tigen. Deshalb mÃ¼ssen Sie den
Zuweisungsoperator (=) benutzen, um den Ausgabewert der open-Methode in einer
Variablen zu speichern:

```python
f = open("willkommen.txt", "r")
print(f.read())
f.close()
```

```python
f = open("hallo.txt", "w")
f.write("Ja Hallo aber auch")
f.close()

f = open("hallo.txt","r")
print(f.read())
f.close()
```

Die oben gezeigte Vorgehensweise ist leider nicht sehr fehlertollerant, demensprechend ist es notwendig eine Fehlerbehandlung vorzusehen. (Kleiner Vorgriff auf das Thema Fehlerbehadlung). Insbesondere das "Vergessen" des f.close() kann zu ernsten Problemen fÃ¼hren, das jede offene Datei ressourcen des Betriebssystem verbraucht (auch wenn das eigentliche Pythonprogramm lÃ¤ngst terminiert ist.

```python
# Klassische Fehlerbehandlung
file = open('hellotest.txt', 'w')
try:
    file.write('hello world')
finally:
    file.close()
```

```python
# mit dem with statement
with open('file_path', 'w') as file:
    file.write('hello world !')
```

Das with Statement kÃ¼mmert sich auch automatisch um das Schliessen der Datei.
